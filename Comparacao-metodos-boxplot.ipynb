## n = 6

Fonte dos dados: [wikipedia.org](https://en.wikipedia.org/wiki/Quartile#Example_2)

| n | Dados ordenados |
| :-: | :-: |
| 1 | 7 |
| 2 | 15 |
| 3 | 36 |
| 4 | 39 |
| 5 | 40 |
| 6 | 41 |


Resumo das estatísticas dos quartis para cada método:


| - | matplotlib | Tukey | M&M | QUARTIL.INC | QUARTIL.EXC |
| :-: | :-: | :-: | :-: | :-: | :-: |
| Mínimo | 7 | 7 | 7 | 7 | 7 |
| $Q1$ | 20.25 | 15 | 15 | 20.25 | 13 | 
| $M_d$ | 37.5 | 37.5 | 37.5 | 37.5 | 37.5 | 
| $Q3$ | 39.75 | 40 | 40 | 39.75 | 40.25 | 
| Máximo | 41 | 41 | 41 | 41 | 41 |
| $DI$ | 19.5 | 25 | 25 | 19.5 | 27.25 |
| $Q1 - 1.5\times DI$ | -9 | -22.5 | -22.5 | -9 | -27.875 |
| $Q3 + 1.5\times DI$ | 69 | 77.5 | 77.5 | 69 | 81.125 |
| Outliers? | No | No | No | No | No |

```python
data = [7, 15, 36, 39, 40, 41]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot() 

## Python
boxplot_data_python = cbook.boxplot_stats(data)[0]
boxplot_data_python['label'] = 'matplotlib'

## Método de Tukey
boxplot_data_tukey = cbook.boxplot_stats(data)[0]
boxplot_data_tukey['q1'] = 15
boxplot_data_tukey['med'] = 37.5
boxplot_data_tukey['q3'] = 40
boxplot_data_tukey['iqr'] = 40-15
boxplot_data_tukey['whislo'] = min(data)
boxplot_data_tukey['whishi'] = max(data)
boxplot_data_tukey['label'] = 'Tukey'

## Método de M & M 
boxplot_data_M_M = cbook.boxplot_stats(data)[0]
boxplot_data_M_M['q1'] = 15
boxplot_data_M_M['med'] = 37.5
boxplot_data_M_M['q3'] = 40
boxplot_data_M_M['iqr'] = 40-15
boxplot_data_M_M['whislo'] = min(data)
boxplot_data_M_M['whishi'] = max(data)
boxplot_data_M_M['label'] = 'M&M'

## Método do Excel QUARTIL.INC()
boxplot_data_INC = cbook.boxplot_stats(data)[0]
boxplot_data_INC['q1'] = 20.25
boxplot_data_INC['med'] = 37.5
boxplot_data_INC['q3'] = 39.75
boxplot_data_INC['iqr'] = 39.75 - 20.25
boxplot_data_INC['whislo'] = min(data)
boxplot_data_INC['whishi'] = max(data)
boxplot_data_INC['label'] = "QUARTIL.INC()"

## Método do Excel QUARTIL.EXC()
boxplot_data_EXC = cbook.boxplot_stats(data)[0]
boxplot_data_EXC['q1'] = 13
boxplot_data_EXC['med'] = 37.5
boxplot_data_EXC['q3'] = 40.25
boxplot_data_EXC['iqr'] = 40.25-13
boxplot_data_EXC['whislo'] = min(data)
boxplot_data_EXC['whishi'] = max(data)
boxplot_data_EXC['label'] = 'QUARTIL.EXC()'


# plotando o boxplot
ax.bxp([boxplot_data_python, boxplot_data_tukey, boxplot_data_M_M, boxplot_data_INC, boxplot_data_EXC])
plt.show()
```
